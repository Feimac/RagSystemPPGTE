```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

# Supondo que 'dados' seja o seu DataFrame carregado
# dados = pd.read_csv('seu_arquivo.csv') # Se seus dados estiverem em um arquivo CSV

# Filtrar os dados para incluir apenas 'ATRASADO' e 'NO HORÁRIO'
dados = dados1[dados1['SITUACAO'].isin(['ATRASADO', 'NO HORÁRIO'])].copy()

# Remover linhas com valores faltantes
dados = dados.dropna().reset_index(drop=True)

# Amostrar uma fração dos dados (por exemplo, 50%)
dados_amostra = dados.sample(frac=0.5, random_state=42) # Ajuste a fração conforme necessário

# Criar a variável dependente binária 'IS_ATRASADO'
dados_amostra['IS_ATRASADO'] = (dados_amostra['SITUACAO'] == 'ATRASADO').astype(int)

# Supondo que 'dados_amostra' seja o seu DataFrame carregado e pré-processado conforme os passos anteriores
# Selecionar as variáveis independentes iniciais
features = ['CODIGOLINHA', 'SENTIDO_IN_TIME', 'LAT_IN_TIME', 'LON_IN_TIME']
X = dados_amostra[features].copy()
y = dados_amostra['IS_ATRASADO']

# Tratar variáveis categóricas com one-hot encoding
X = pd.get_dummies(X, columns=['CODIGOLINHA','SENTIDO_IN_TIME'], drop_first=True)

# Dividir os dados em conjunto de treinamento e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Criar e treinar o modelo de regressão logística
model = LogisticRegression(solver='liblinear', random_state=42)
model.fit(X_train, y_train)
print("Modelo de Regressão Logística treinado com sucesso com dados amostrados!")
```

```python
y_pred_proba = model.predict_proba(X_test)[:, 1]
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Calcular as métricas
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)

print(f'Acurácia: {accuracy:.4f}')
print(f'Precisão: {precision:.4f}')
print(f'Recall: {recall:.4f}')
print(f'F1-Score: {f1:.4f}')

# Visualizar a matriz de confusão
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Não Atrasado', 'Atrasado'],
            yticklabels=['Não Atrasado', 'Atrasado'])
plt.xlabel('Previsto')
plt.ylabel('Real')
plt.title('Matriz de Confusão')
plt.show()
```

```python
from sklearn.metrics import roc_curve, roc_auc_score

fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)
auc = roc_auc_score(y_test, y_pred_proba)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'AUC = {auc:.2f}')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlabel('Taxa de Falsos Positivos (FPR)')
plt.ylabel('Taxa de Verdadeiros Positivos (TPR)')
plt.title('Curva ROC')
plt.legend(loc='lower right')
plt.show()

print(f'AUC: {auc:.2f}')
```

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Carregar os dados
dados1 = pd.read_csv('veiculos.csv')

# Filtrar os dados para incluir apenas 'ATRASADO' e 'NO HORÁRIO'
dados = dados1[dados1['SITUACAO'].isin(['ATRASADO', 'NO HORÁRIO'])].copy()

# Remover linhas com valores faltantes
dados = dados.dropna().reset_index(drop=True)

# Criar a variável dependente binária 'IS_ATRASADO'
dados['IS_ATRASADO'] = (dados['SITUACAO'] == 'ATRASADO').astype(int)

# Calcular as frequências de CODIGOLINHA vs IS_ATRASADO para o grafo
freq_linha_atraso = dados.groupby(['CODIGOLINHA', 'IS_ATRASADO']).size().unstack(fill_value=0)
freq_sentido_atraso = dados.groupby(['SENTIDO_IN_TIME', 'IS_ATRASADO']).size().unstack(fill_value=0)

# Criar o grafo
G_comparacao = nx.Graph()

# Adicionar nós para as categorias de CODIGOLINHA e IS_ATRASADO
for linha in freq_linha_atraso.index:
    G_comparacao.add_node(f"Linha_{linha}", type="CODIGOLINHA")
G_comparacao.add_node("Nao_Atrasado", type="IS_ATRASADO")
G_comparacao.add_node("Atrasado", type="IS_ATRASADO")

# Adicionar arestas baseadas na frequência de CODIGOLINHA vs IS_ATRASADO
for linha, row in freq_linha_atraso.iterrows():
    nao_atrasado_count = row.get(0, 0)
    atrasado_count = row.get(1, 0)
    if nao_atrasado_count > 0:
        G_comparacao.add_edge(f"Linha_{linha}", "Nao_Atrasado", weight=nao_atrasado_count)
    if atrasado_count > 0:
        G_comparacao.add_edge(f"Linha_{linha}", "Atrasado", weight=atrasado_count)

# Adicionar nós para as categorias de SENTIDO_IN_TIME e IS_ATRASADO
for sentido in freq_sentido_atraso.index:
    G_comparacao.add_node(f"Sentido_{sentido}", type="SENTIDO_IN_TIME")

# Adicionar arestas baseadas na frequência de SENTIDO_IN_TIME vs IS_ATRASADO
for sentido, row in freq_sentido_atraso.iterrows():
    nao_atrasado_count = row.get(0, 0)
    atrasado_count = row.get(1, 0)
    if nao_atrasado_count > 0:
        G_comparacao.add_edge(f"Sentido_{sentido}", "Nao_Atrasado", weight=nao_atrasado_count)
    if atrasado_count > 0:
        G_comparacao.add_edge(f"Sentido_{sentido}", "Atrasado", weight=atrasado_count)

# Visualizar o Grafo (opcional - pode ser complexo para muitos dados)
plt.figure(figsize=(12, 8))
pos = nx.spring_layout(G_comparacao, k=0.8)
node_colors = ['lightblue' if data['type'] == "CODIGOLINHA" else 'lightgreen' if data['type'] == "SENTIDO_IN_TIME" else 'salmon' for _, data in G_comparacao.nodes(data=True)]
edge_widths = [d['weight'] * 0.005 for u, v, d in G_comparacao.edges(data=True)] # Ajuste a escala conforme necessário
nx.draw(G_comparacao, pos, with_labels=True, node_color=node_colors, width=edge_widths, node_size=2000, alpha=0.7)
plt.title("Relação entre Variáveis Independentes e Atraso (Frequência)")
plt.show()

# Preparar dados para regressão (usando as mesmas features da regressão anterior)
features = ['CODIGOLINHA', 'SENTIDO_IN_TIME', 'LAT_IN_TIME', 'LON_IN_TIME']
X = dados[features].copy()
y = dados['IS_ATRASADO']

# Tratar variáveis categóricas com one-hot encoding
X = pd.get_dummies(X, columns=['CODIGOLINHA', 'SENTIDO_IN_TIME'], drop_first=True)

# Dividir os dados em conjunto de treinamento e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Criar e treinar o modelo de regressão logística
model = LogisticRegression(solver='liblinear', random_state=42)
model.fit(X_train, y_train)
print("\nModelo de Regressão Logística treinado com sucesso!")

# Avaliar o modelo
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f'Acurácia da Regressão Logística: {accuracy:.4f}')
print(f'Precisão da Regressão Logística: {precision:.4f}')
print(f'Recall da Regressão Logística: {recall:.4f}')
print(f'F1-Score da Regressão Logística: {f1:.4f}')

# --- Aplicando informações do grafo como features (exemplo: proporção de atrasos por linha) ---

proporcao_atraso_linha = {}
for linha in freq_linha_atraso.index:
    total = freq_linha_atraso.loc[linha, 0] + freq_linha_atraso.loc[linha, 1]
    if total > 0:
        proporcao = freq_linha_atraso.loc[linha, 1] / total
        proporcao_atraso_linha[linha] = proporcao

# Criar uma nova feature no DataFrame original com a proporção de atrasos da linha
dados['PROPORCAO_ATRASO_LINHA'] = dados['CODIGOLINHA'].map(proporcao_atraso_linha).fillna(0)

# Preparar os dados para uma nova regressão com a feature do grafo
features_com_grafo = ['CODIGOLINHA', 'SENTIDO_IN_TIME', 'LAT_IN_TIME', 'LON_IN_TIME', 'PROPORCAO_ATRASO_LINHA']
X_com_grafo = dados[features_com_grafo].copy()
y_com_grafo = dados['IS_ATRASADO']

# Tratar variáveis categóricas com one-hot encoding
X_com_grafo = pd.get_dummies(X_com_grafo, columns=['CODIGOLINHA', 'SENTIDO_IN_TIME'], drop_first=True)

# Dividir os dados em conjunto de treinamento e teste
X_train_grafo, X_test_grafo, y_train_grafo, y_test_grafo = train_test_split(X_com_grafo, y_com_grafo, test_size=0.3, random_state=42)

# Criar e treinar o modelo de regressão logística com a feature do grafo
model_com_grafo = LogisticRegression(solver='liblinear', random_state=42)
model_com_grafo.fit(X_train_grafo, y_train_grafo)
print("\nModelo de Regressão Logística treinado com feature do grafo!")

# Avaliar o modelo com a feature do grafo
y_pred_grafo = model_com_grafo.predict(X_test_grafo)
accuracy_grafo = accuracy_score(y_test_grafo, y_pred_grafo)
precision_grafo = precision_score(y_test_grafo, y_pred_grafo)
recall_grafo = recall_score(y_test_grafo, y_pred_grafo)
f1_grafo = f1_score(y_test_grafo, y_pred_grafo)

print(f'Acurácia da Regressão Logística (com feature do grafo): {accuracy_grafo:.4f}')
print(f'Precisão da Regressão Logística (com feature do grafo): {precision_grafo:.4f}')
print(f'Recall da Regressão Logística (com feature do grafo): {recall_grafo:.4f}')
print(f'F1-Score da Regressão Logística (com feature do grafo): {f1_grafo:.4f}')
```
